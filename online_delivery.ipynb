```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Zomato dataset
zomato_data = pd.read_csv("/content/zomato.csv")

# Filter restaurants that offer online delivery and those that don't
online_delivery_restaurants = zomato_data[zomato_data['online_order'] == 'Yes']
no_online_delivery_restaurants = zomato_data[zomato_data['online_order'] == 'No']

# Calculate the average rating of restaurants with online delivery and without online delivery
average_rating_with_online_delivery = online_delivery_restaurants['rate (out of 5)'].mean()
average_rating_without_online_delivery = no_online_delivery_restaurants['rate (out of 5)'].mean()

# Create a bar graph to compare the average ratings of restaurants with and without online delivery
plt.figure(figsize=(8, 6))
plt.bar(['With Online Delivery', 'Without Online Delivery'], [average_rating_with_online_delivery, average_rating_without_online_delivery], color=['skyblue', 'lightcoral'])
plt.title('Comparison of Average Ratings')
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.ylim(0, 5)  # Set y-axis limit to 0-5 for rating scale
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print the restaurants that offer online delivery
print("Restaurants that offer online delivery:")
print(online_delivery_restaurants)

# Print the restaurants that don't offer online delivery
print("\nRestaurants that don't offer online delivery:")
print(no_online_delivery_restaurants)

# Print the average ratings of restaurants with and without online delivery
print("\nAverage rating of restaurants with online delivery:", average_rating_with_online_delivery)
print("Average rating of restaurants without online delivery:", average_rating_without_online_delivery)
```
